```python
import pandas as pd
import matplotlib.pyplot as plt

# Create example data
data = [
    {"name": "Alice", "unitprice": 10, "quantity": 2, "date": "2024-01-01"},
    {"name": "Bob", "unitprice": 15, "quantity": 1, "date": "2024-01-02"},
    {"name": "Alice", "unitprice": 12, "quantity": 3, "date": "2024-01-03"},
    {"name": "Charlie", "unitprice": 8, "quantity": 5, "date": "2024-01-01"},
    {"name": "Bob", "unitprice": 20, "quantity": 2, "date": "2024-01-05"}
]

# Convert to DataFrame
df = pd.DataFrame(data)

# Save to CSV
df.to_csv('SalesData.csv', index=False)

# Read CSV into DataFrame
df = pd.read_csv('SalesData.csv')
print("Phase 1 - Data read successfully:")
print(df)
```

```python
print("\nPhase 2 - Description of 'unitprice':")
print(df['unitprice'].describe())
```

```python
# Calculate net_price
df['net_price'] = df['unitprice'] * df['quantity']

# Create new dataframe with required columns
df_filtered = df[['name', 'net_price', 'date']]

print("\nPhase 3 - Filtered DataFrame:")
print(df_filtered)

# Group by customer name (sum of net_price)
df_grouped = df_filtered.groupby('name', as_index=False).sum()
print("\nGrouped total sales by customer:")
print(df_grouped)
```

```python
plt.figure(figsize=(8,5))

# Bar plot: Customer name vs Total sales
plt.bar(df_grouped['name'], df_grouped['net_price'], color='skyblue')

plt.xlabel('Customer Name')
plt.ylabel('Total Sales ($)')
plt.title('Total Sales by Each Customer')
plt.show()
```

```python
import pandas as pd

# Create dataframe
data = {
    'first_name': ['Jason', 'Molly', 'Tina', 'Jake', 'Amy'],
    'last_name': ['Miller', 'Jacobson', 'Ali', 'Milner', 'Cooze'],
    'female': [0, 1, 1, 0, 1],
    'age': [42, 52, 36, 24, 73],
    'preTestScore': [4, 24, 31, 2, 3],
    'postTestScore': [25, 94, 57, 62, 70]
}

df = pd.DataFrame(data)

# Scatter plot: size determined by age
plt.scatter(df['preTestScore'], df['postTestScore'], s=df['age']*10, c='orange', alpha=0.6)
plt.xlabel("Pre Test Score")
plt.ylabel("Post Test Score")
plt.title("Pre vs Post Test Scores with Age Size")
plt.show()
```

```python
# Scatter plot with fixed size and color based on sex
plt.figure(figsize=(6,5))

# Map colors: 0 -> blue, 1 -> pink
colors = df['female'].map({0: 'blue', 1: 'pink'})

plt.scatter(df['preTestScore'], df['postTestScore'], s=300, c=colors, alpha=0.7)
plt.xlabel("Pre Test Score")
plt.ylabel("Post Test Score")
plt.title("Pre vs Post Test Scores Colored by Sex")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
df = pd.read_csv("BigMartSalesData.csv")

# Check the first few rows
df.head()
```

```python
# Convert InvoiceDate to datetime
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])

# Extract month
df['Month'] = df['InvoiceDate'].dt.month

# Group by month and calculate total sales
monthly_sales = df.groupby('Month')['Amount'].sum()

# Plot line chart
plt.figure(figsize=(8,5), dpi=100)
plt.plot(monthly_sales.index, monthly_sales.values, marker='o', linestyle='-', color='green')
plt.title("Total Sales Per Month - 2011", fontsize=14)
plt.xlabel("Month")
plt.ylabel("Total Sales")
plt.xticks(range(1,13))
plt.grid(True)
plt.show()

# Identify the month with lowest sales
lowest_month = monthly_sales.idxmin()
lowest_sales = monthly_sales.min()
print(f"Month with lowest sales: {lowest_month}, Sales: {lowest_sales}")
```

```python
# Bar Chart
plt.figure(figsize=(8,5), dpi=100)
plt.bar(monthly_sales.index, monthly_sales.values, color='skyblue')
plt.title("Total Sales Per Month - 2011 (Bar Chart)", fontsize=14)
plt.xlabel("Month")
plt.ylabel("Total Sales")
plt.xticks(range(1,13))
plt.show()
```

```python
# Group by country
country_sales = df.groupby('Country')['Amount'].sum()

# Pie chart
plt.figure(figsize=(7,7), dpi=100)
plt.pie(country_sales, labels=country_sales.index, autopct='%1.1f%%', startangle=140, shadow=True)
plt.title("Country Wise Sales Distribution - 2011", fontsize=14)
plt.show()

# Identify country with highest sales
top_country = country_sales.idxmax()
top_sales = country_sales.max()
print(f"Country with highest sales: {top_country}, Sales: {top_sales}")
```

```python
plt.figure(figsize=(10,5), dpi=100)

# Scatter plot: x = invoice index, y = invoice amount
plt.scatter(df.index, df['Amount'], color='blue', alpha=0.5)

plt.title("Scatter Plot of Invoice Amounts - 2011", fontsize=14)
plt.xlabel("Invoice Number")
plt.ylabel("Invoice Amount (Sales)")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8,5), dpi=100)

plt.hist(df['Amount'], bins=30, color='orange', edgecolor='orange')
plt.title("Distribution of Invoice Amounts - 2011", fontsize=14)
plt.xlabel("Invoice Amount (Sales)")
plt.ylabel("Number of Invoices")
plt.show()
```
