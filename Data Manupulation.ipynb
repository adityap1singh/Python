```python
import pandas as pd
# noinspection PyUnusedImports
import matplotlib.pyplot as plt

df = pd.read_csv("HollywoodMovies.csv")
```

```python
quest_movies = df[df["story"] == "AudienceScore"]
highest_rated_quest = quest_movies.loc[
    quest_movies["RottenTomatoes"].idxmax()
]

highest_rated_quest[["movie_name", "rotten_tomatoes_rating"]]
```

```python
most_common_genre = df["Genre"].value_counts().idxmax()
most_common_genre
```

```python
top_budget_movies = df.sort_values(
    by="Budget", ascending=False
).head(5)

top_budget_movies[["Movie", "Budget"]]
```

```python
df["Openprofit"] = df["Profitability"] - df["Budget"]
```

```python
plt.figure()
plt.scatter(df["Rottentomatoes"], df["OpenProfit"])
plt.xlabel("RottenTomatoes Rating")
plt.ylabel("Profitability")
plt.title("Critics Rating vs Public Acceptance")
plt.show()
```

```python
import pandas as pd
import numpy as np

df = pd.read_csv("Salaries.csv")

# Replace NaN with 0 for pay-related columns
pay_cols = ['BasePay', 'OvertimePay', 'OtherPay', 'TotalPay', 'TotalPayBenefits']
df[pay_cols] = df[pay_cols].fillna(0)
```

```python
salary_2011 = df[df['Year'] == 2011]['TotalPayBenefits'].sum()
salary_2014 = df[df['Year'] == 2014]['TotalPayBenefits'].sum()

increase = salary_2014 - salary_2011
increase
```

```python
df_2014 = df[df['Year'] == 2014]

highest_mean_job = (
    df_2014.groupby('JobTitle')['TotalPayBenefits']
    .mean()
    .idxmax()
)

highest_mean_job
```

```python
overtime_savings_2014 = df_2014['OvertimePay'].sum()
overtime_savings_2014
```

```python
top5_jobs = df_2014['JobTitle'].value_counts().head(5)

top5_cost = (
    df_2014[df_2014['JobTitle'].isin(top5_jobs.index)]
    .groupby('JobTitle')['TotalPayBenefits']
    .sum()
)

top5_cost
```

```python
top_earner = df.groupby('EmployeeName')['TotalPayBenefits'].sum().idxmax()
top_earner
```

```python
last5_jobs = df_2014['JobTitle'].value_counts().tail(5)

last5_cost = (
    df_2014[df_2014['JobTitle'].isin(last5_jobs.index)]
    .groupby('JobTitle')['TotalPayBenefits']
    .sum()
)

last5_cost
```

```python
overtime_percentage = (
    df_2014['OvertimePay'].sum() /
    df_2014['TotalPayBenefits'].sum()
) * 100

overtime_percentage
```

```python
lowest_mean_job = (
    df_2014.groupby('JobTitle')['TotalPayBenefits']
    .mean()
    .idxmin()
)

lowest_mean_job
```
